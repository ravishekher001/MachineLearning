```python
import bs4
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://www.gutenberg.org/files/15659/15659-h/15659-h.htm"
```

```python
r = requests.get(url)
```

```python
html_doc = r.text
```

```python
soup = BeautifulSoup(html_doc)
```

```python
pretty_soup = soup.prettify()
```

```python
print(pretty_soup[:1000])
```

```python
print(soup.get_text())
```
